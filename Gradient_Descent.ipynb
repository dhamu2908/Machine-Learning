```python
#Data
X = [0.3, 1.7]
Y = [0.1, 0.8]
```

```python
print(X)
print(Y)
```

```python
import numpy as np
```

```python
#Logistic Function (Sigmoid Function)
def f(x,w,b): 
  return 1/(1+np.exp(-(w*x+b)))
```

```python
#Error Function MSE
def error(w,b):
  err = 0.0
  for x,y in zip(X,Y):
    fx = f(x,w,b)
    err += (fx-y)**2
  return (1/len(X))*err
```

```python
#Derivative of gradient w.r.t bias term
def grad_b(x,w,b,y):
  fx = f(x,w,b)
  return (fx-y)*fx*(1-fx)

#Derivative of gradient w.r.t weights
def grad_w(x,w,b,y):
  fx = f(x,w,b)
  return (fx-y)*fx*(1-fx)*x
```

```python
def do_gradient_descent():
  
  w,b,eta,max_epochs = -1,-2,0.5,1000

  print(f"Initially w : {w} , b : {b}")
  
  for epoch in range(max_epochs):
    dw,db = 0,0
    for x,y in zip(X,Y):
      dw += grad_w(x,w,b,y)
      db += grad_b(x,w,b,y)
    
    w_new = w - eta*dw
    b_new = b - eta*db

    print(f"After epoch : {epoch + 1}  ,   w : {w_new:.5f}   ,   b : {b_new:.5f}    ,   error : {error(w_new,b_new):.5f}")

    if w_new - w == 0 and b_new - b == 0:
        print(f"Converged in {epoch + 1} iterations")
        break
    w = w_new
    b = b_new
        
  print(f"Final parameters : w = {w_new:.5f} , b = {b_new:.5f}")
  print(f"Final error : {error(w_new,b_new):.5f}")
  print(f"Converged in {epoch + 1} iterations")

        
```

```python
do_gradient_descent()
```
