```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
print(x)
print(y)
print(len(y))
```

```python
print(np.unique(y))
```

```python

def gaussian_pdf(x,mean,std):
  eps = 1e-9
  std = np.maximum(std,eps)
  exponent = np.exp(-0.5*((x-mean)/std)**2)
  return (1/(np.sqrt(2*np.pi)*std))*exponent


def naive_bayes(X,classes,cls_prior,feature_stats):
  y_pred = []
  for x in X:
    cls_prob = {}
    for i,cls in enumerate(classes):
      log_prior = np.log(cls_prior[i])
      log_likelihood = np.sum(np.log(gaussian_pdf(x,feature_stats[cls]["mean"],feature_stats[cls]["std"])))
      cls_prob[cls] = log_prior + log_likelihood
    y_pred.append(max(cls_prob,key=cls_prob.get))
  return y_pred

classes = np.unique(y)

cls_prior = []

feature_stats = {}

for cls in classes:
  x_cls = x[y==cls]
  cls_prior.append(len(x_cls)/len(x))
  feature_stats[cls] = {
            "mean": np.mean(x_cls, axis=0),
            "std": np.std(x_cls, axis=0)
        }
print(cls_prior)
print(feature_stats)

y_pred = naive_bayes(x,classes,cls_prior,feature_stats)
print("Y pred : ",np.array(y_pred))
print()
print(y)
```
