```python
#Simple Linear Regression
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Data
x = np.array([5,9,10,12,16])
y = np.array([2, 4, 6, 6, 7])

n=len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)


numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)

slope = numerator/denominator
intercept = y_mean - slope * x_mean

```

```python
#slope
print("Slope = ",end='')
print(slope)

#Intercept
print("Intercept = ",end='')
print(intercept)

# Plot the data and regression line
plt.scatter(x, y)
plt.plot(x, slope * x + intercept, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.show()
```

```python
#Prediction for new data

# y = m*x + c

# slope = m
# Intercept = c

#For new data point x
new_x=16

print(slope * new_x + intercept)
```

```python
#Computing Parameters using GRADIENT DESCENT
```

```python
#Hyperparameters

learning_rate = 0.001
epochs = 5

m=0  # Slope
c=0  # Intercept

for epoch in range(epochs):
  y_pred = m * x + c

  #Calculating the gradients
  #∂MSE/∂m
  #∂MSE/∂c

  dm = (-2/n) * np.sum(x * (y - y_pred))
  dc = (-2/n) * np.sum(y - y_pred)

  # Update parameters
  m -= learning_rate * dm
  c -= learning_rate * dc

  print(f"Epoch {epoch + 1}: m = {m:.4f}, c = {c:.4f}")

# Plotting the data points
plt.scatter(x, y, color='blue', label='Data_Points')

# Plotting the regression line
plt.plot(x, m * x + c, color='red', label=f'Regression Line')

# Labeling the axes
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression using MSE')

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```

```python
#Mean Squared Error
```

```python
MSE = 0
for i in range(n):
  y_pred = m * x[i] + c
  MSE += ( y[i] - y_pred ) ** 2

print("Mean Squared Error = ", end=' ')
print(MSE)
```
