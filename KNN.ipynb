```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("train.csv")
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

print(x)
print(y)
print(len(y))
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))


def predict(X_train, y_train, X_test, k):
  predictions = []
  for x in X_test:
    distances = [euclidean_distance(x, x_train) for x_train in X_train]
    k_indices = np.argsort(distances)[:k]
    k_labels = [y_train[i] for i in k_indices]
    predictions.append(max(k_labels, key=k_labels.count))  # Most common label
  return predictions
```

```python
y_pred =  predict(x, y, x, k=1)

print(y_pred)
print(y)
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_boundaries(X_train, y_train, k, resolution=0.1):
    # Create a color map
    cmap_light = ListedColormap(["#FFAAAA", "#AAFFAA", "#AAAAFF"])
    cmap_bold = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])

    # Create a mesh grid
    x_min, x_max = X_train[:, 0].min() - 1, X_train[:, 0].max() + 1
    y_min, y_max = X_train[:, 1].min() - 1, X_train[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, resolution),
                         np.arange(y_min, y_max, resolution))

    # Predict the label for each point in the grid
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    Z = predict(X_train, y_train, grid_points, k)
    Z = np.array(Z).reshape(xx.shape)

    # Plot the decision boundary
    plt.figure(figsize=(10, 6))
    plt.contourf(xx, yy, Z, alpha=2, cmap=cmap_light)

    # Plot the training points
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cmap_bold, edgecolor='k', s=50)
    plt.title(f"KNN Decision Boundary (k={k})")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.show()
```

```python
plot_decision_boundaries(x,y,4)
```
